```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Value-oriented consumers - Sheet1.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['ICE_Total_Cost'], color='red', label='ICE TCO', kde=True)
sns.histplot(df['EV_Total_Cost'], color='green', label='EV TCO', kde=True)
plt.title('Total Cost of Ownership: ICE vs EV')
plt.xlabel('TCO (â‚¹)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Aware_Govt_Incentives', data=df)
plt.title('Awareness of Government Incentives for EVs')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Prefers_EV', y='Cost_Sensitivity', data=df)
plt.title('Cost Sensitivity vs Preference for EVs')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Location', hue='Prefers_EV', data=df)
plt.title('Location-wise Preference for EVs')
plt.show()
```
